```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf
from cgi import print_exception
from operator import contains

date = "5/5/2023"

response = requests.get("https://eresearch.fidelity.com/eresearch/conferenceCalls.jhtml?tab=dividends&begindate="+date)

content = response.content
content.count()
parser = BeautifulSoup(content, 'html.parser')
first_paragraph = parser.find_all("table", id="messageTable3")[0]
data_frame = pd.read_html(str(first_paragraph))[0]


data = data_frame.drop(['Anouncement Date','Record Date'],axis=1)
data = data.sort_values(by='Dividend',ascending=False)
data.head(10)
```

```python

def getLastPrice(symbol):
    #symbol = "GSK"
    url = "https://eresearch.fidelity.com/eresearch/evaluate/snapshot.jhtml?symbols="+symbol
    response = requests.get(url)
    if response.ok :
        try:
            parser = BeautifulSoup(response.content, 'html.parser')
            return float(parser.find_all("span", id="lastPrice")[0].contents[0])
        except:
            print_exception
    
```

```python
data = data.drop_duplicates('Symbol')
keys = data['Symbol']
```

```python

for i in keys:
    #caution possible different stocks symbol mix ':CA' etc
    if contains(i,':'):
      j  = (i.split(sep=":"))[0]
    else:
      j=i
    price = getLastPrice(j)
    data.loc[data['Symbol'] == i, 'Price'] = price
    #print(i)
```

```python
try:
    data['DivYield'] = data['Dividend'] / data['Price']
    data = data.sort_values(by='DivYield',ascending=False)
except:
    print_exception
data
```
