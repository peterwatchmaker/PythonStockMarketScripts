```python
from datetime import datetime, timedelta
date = datetime.today()
datestr = date.date().strftime("%m/%d/%Y")
def getEarningsForDate(date):
    response = requests.get("https://eresearch.fidelity.com/eresearch/conferenceCalls.jhtml?begindate="+date)
    content = response.content
    parser = BeautifulSoup(content, 'html.parser')
    data_frame = pd.read_html(str(parser.contents[3].body.table),header=1)[0]
    data_frame['ADate'] = date
    return data_frame

data = pd.DataFrame()
for i in range(0,7):
    date = date + timedelta(i)
    datestr = date.date().strftime("%m/%d/%Y")
    if not data.empty:
        data.join(getEarningsForDate(datestr),how='outer',lsuffix="_L_",rsuffix="_R_")
    else:
        data = getEarningsForDate(datestr)
    #data.append(getEarningsForDate(datestr))

data = data.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'],axis=1)
#data.loc[data['Symbol'] == 'AGFY']
data.loc[data['Time of Day'] == 'Before Market']
```
